### NumPy – Matrix Operations, Loops & Conditions

```python
'''Why?
NumPy is a foundational package for scientific computing. It supports large, multi-dimensional arrays and matrices along with mathematical functions to operate on them.

Functions Used:

np.array() – creates an array

np.mean() – calculates the mean

np.median() – calculates the median

np.var() – variance

np.std() – standard deviation

Where Used?
To do basic statistics like mean, variance, etc., on a dataset.'''
```

```python
import numpy as np

# Create 2D arrays
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 3]])

# Matrix operations
print("Addition:\n", A + B)
print("Dot Product:\n", np.dot(A, B))
print("Element-wise multiplication:\n", A * B)

# Loop through array
for row in A:
    print("Row:", row)

# Conditional operation
print("Elements greater than 2:\n", A[A > 2])
```

### Matplotlib – Plotting

```python
'''Why?
Matplotlib is a plotting library used for creating static, animated, and interactive visualizations in Python.

Functions Used:

plt.plot() – line plot

plt.bar() – bar plot

plt.hist() – histogram

plt.xlabel(), plt.ylabel(), plt.title() – for labeling plots

plt.show() – render the plot

Where Used?
To visualize data (e.g., plotting a bar chart or histogram of a variable's distribution).'''
```

```python
import matplotlib.pyplot as plt

# Line plot
x = [1, 2, 3, 4, 5]
y = [10, 12, 15, 18, 20]
plt.plot(x, y, marker='o', color='blue')
plt.title("Line Chart Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()

# Scatter plot from Iris
plt.scatter(df["sepal_length"], df["sepal_width"], c="green", alpha=0.6)
plt.title("Sepal Length vs Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

### Statistics – Mean, Median, Mode, Variance, Std. Dev.

```python
'''Why?
SciPy builds on NumPy and provides functions for statistics, optimization, and more.

Functions Used:

stats.mode() – returns the mode (most frequent value)

Where Used?
To find the mode in a list of values (since NumPy does not have a built-in mode()).'''
```

```python
import numpy as np
from scipy import stats

# Sample list
data = [12, 15, 21, 15, 19, 15, 25, 30, 12]

# Statistical measures
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Mode:", stats.mode(data).mode[0])
print("Variance:", np.var(data))
print("Standard Deviation:", np.std(data))
```

```python
'''What is the Iris Dataset?
The Iris dataset contains 150 samples of iris flowers from three different species:

Setosa

Versicolor

Virginica

Each sample includes four numerical features (also called attributes):

Feature	Description	Unit
sepal length (cm)	Length of the sepal	centimeters
sepal width (cm)	Width of the sepal	centimeters
petal length (cm)	Length of the petal	centimeters
petal width (cm)	Width of the petal	centimeters

And the target variable (label) is the species of the flower.'''
```

```python
'''Dataset Structure
The dataset has:

150 rows (flowers)

4 columns of features

1 column of label (the species)

Why is it Popular?
It is small (easy to load into memory).

It has no missing values.

The data is well-labeled.

It’s a multi-class classification problem (3 classes).

Useful for:

Data visualization

Feature selection

Classification model building

Clustering (unsupervised learning)

'''
```

```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the dataset
iris = load_iris()

# Convert to DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['species'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Show the first 5 rows
print(df.head())
```

```python
'''This script gives you a quick and powerful visual summary of:

The distribution of each feature.

The correlation between features.

The separability of the classes (species).'''
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, hue='species')
plt.show()
```

```python
'''
import seaborn as sns
Seaborn is a Python data visualization library based on Matplotlib.

It provides a high-level interface for drawing attractive and informative statistical graphics.

2. import matplotlib.pyplot as plt
Matplotlib is the foundational plotting library in Python.

pyplot (aliased as plt) allows you to create plots and charts like line graphs, bar graphs, etc.

Seaborn builds on top of this.

3. sns.pairplot(df, hue='species')
This creates a pair plot: a grid of scatterplots showing relationships between every pair of numerical columns in the DataFrame.

df is your DataFrame (assumed to be the Iris dataset in this case).

hue='species' means the points in the plots will be color-coded based on flower species.

This helps us visually identify patterns, like:

Which features best separate different flower species?

Are there clusters of similar data points?

Example output:

On the diagonal: histograms of each feature.

Below/above the diagonal: scatterplots of each feature pair.

4. plt.show()
This tells Matplotlib to display the plots.

Without this, the plots may not appear (especially in some environments like scripts or IDEs).

'''
```
